```python
import pandas as pd

# URL da planilha do Google Sheets (somente ID)
sheet_id = '10tpwe3_rfH-TZEuXpNT6vCUaqOK1aofZ'

# URL para a planilha no formato CSV
sheet_url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv'

# Carregar os dados da planilha em um DataFrame
df = pd.read_csv(sheet_url)

# Exibir as primeiras linhas do DataFrame
df.head()


```

```python
# NÃºmero de colunas
num_colunas = df.shape[1]
print(f'O nÃºmero de colunas Ã©: {num_colunas}')
```

```python
# NÃºmero de linhas
num_linhas = df.shape[0]
print(f'O nÃºmero de linhas Ã©: {num_linhas}')
```

```python
# Tipos de caracterÃ­sticas
tipos_de_dados = df.dtypes
print(tipos_de_dados)

# Separando as colunas numÃ©ricas e categÃ³ricas
colunas_numericas = df.select_dtypes(include=['number']).columns
colunas_categoricas = df.select_dtypes(exclude=['number']).columns

print(f'CaracterÃ­sticas numÃ©ricas: {colunas_numericas}')
print(f'CaracterÃ­sticas categÃ³ricas: {colunas_categoricas}')
```

```python
# Exibindo estatÃ­sticas descritivas das colunas numÃ©ricas
df.describe()
```

```python
# FrequÃªncia das caracterÃ­sticas categÃ³ricas
for col in colunas_categoricas:
    print(f'\nFrequÃªncia de {col}:')
    print(df[col].value_counts())

```

```python
# Verificando valores faltantes
dados_faltantes = df.isnull().sum()
print(dados_faltantes)
```
